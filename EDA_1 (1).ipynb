```python
sslc["FL"].replace("AA", np.NAN, inplace = True)
sslc["SL"].replace("AA", np.NAN, inplace = True)
sslc["Maths"].replace("AA", np.NAN, inplace = True)
sslc["Science"].replace("AA", np.NAN, inplace = True)
sslc["Social Studies"].replace("AA", np.NAN, inplace = True)
```

```python
sslc["FL"] = pd.to_numeric(sslc["FL"], errors = "coerce")
sslc["SL"] = pd.to_numeric(sslc["SL"], errors = "coerce")
sslc["Maths"] = pd.to_numeric(sslc["Maths"], errors = "coerce")
sslc["Science"] = pd.to_numeric(sslc["Science"], errors = "coerce")
sslc["Social Studies"] = pd.to_numeric(sslc["Social Studies"], errors = "coerce")
sslc["Total"] = pd.to_numeric(sslc["Total"], errors = "coerce")
```
